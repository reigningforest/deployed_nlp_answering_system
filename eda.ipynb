# Loading in the Data

```python
import pandas as pd
import os

def load_messages_to_dataframe(filename="data/all_messages.json"):
    """
    Reads a JSON file containing a list of message objects into a pandas DataFrame.

    The JSON file is expected to be in the 'records' format, which is a
    list of dictionaries.

    Args:
        filename (str): The path to the JSON file.

    Returns:
        pd.DataFrame: A pandas DataFrame containing the message data,
                      or None if the file is not found or is invalid.
    """
    if not os.path.exists(filename):
        print(f"Error: File not found at '{filename}'")
        return None

    try:
        # pd.read_json can directly handle a list of JSON objects
        df = pd.read_json(filename, orient='records')
        print(f"Successfully loaded {len(df)} messages into a DataFrame.")
        return df
    except ValueError as e:
        print(f"Error reading JSON file. It might not be in the expected format: {e}")
        return None
```

```python
df = load_messages_to_dataframe()

df.sort_values(by=['user_name', 'timestamp'], inplace=True)

df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# --- Plotting the Histogram ---
plt.style.use('seaborn-v0_8-whitegrid')
fig, ax = plt.subplots(figsize=(14, 8))

# Create the histogram. You can adjust the number of bins.
ax.hist(df['date'], bins=30, edgecolor='black', color='#4C72B0', alpha=0.7)

# --- Formatting the Plot for Clarity ---
ax.set_title('Distribution of Messages Over Time', fontsize=16, pad=20)
ax.set_xlabel('Date', fontsize=12)
ax.set_ylabel('Number of Messages', fontsize=12)

# Format the x-axis to display dates nicely
fig.autofmt_xdate(rotation=45)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
ax.xaxis.set_major_locator(mdates.MonthLocator(interval=1))
ax.tick_params(axis='x', labelsize=10)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Convert timestamp strings to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])

# --- Feature Engineering (as you provided) ---
df['date'] = df['timestamp'].dt.date
df['hour'] = df['timestamp'].dt.hour
# Classify as 'Day' (6 AM to 5:59 PM) or 'Night'
df['daynight'] = df['hour'].apply(lambda x: 'Day' if 6 <= x < 18 else 'Night')

# --- Data Aggregation ---
# Group by date and day/night category, then count messages
daily_counts = df.groupby(['date', 'daynight']).size().reset_index(name='message_count')

# --- Plotting the Boxplot ---
plt.style.use('seaborn-v0_8-whitegrid')
plt.figure(figsize=(10, 7))

# Create the boxplot using seaborn
sns.boxplot(
    x='daynight',
    y='message_count',
    data=daily_counts,
    palette=['#8ECFC9', '#4C72B0'], # Day/Night colors
    order=['Day', 'Night'] # Ensure consistent order
)

# --- Formatting the Plot for Clarity ---
plt.title('Distribution of Daily Message Counts (Day vs. Night)', fontsize=16, pad=20)
plt.xlabel('Time of Day', fontsize=12)
plt.ylabel('Number of Messages per Day', fontsize=12)
plt.tick_params(labelsize=10)

plt.tight_layout()
plt.show()

```

Average of 5 messages during the day and 4 messages during the night per user.
